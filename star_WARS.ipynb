```python
import scrapy
import requests 
from scrapy.selector import Selector

url = 'http://gif-explode.com/?explode=http://www.python-challenges.com/psb/static/img/asteroid_storm/QJAJQ_OJIJS/FOBCTK_ZSKDK/qPbT.gif'
xpath = '/html/body/div[2]/div/div[4]/img/@src'
req = requests.get(url)
data = Selector(text=req.content.decode()).xpath(xpath).extract()
for item in data:
    print (item)
```

```python
import qrtools
qr = qrtools.QR()
qr.decode(item)
print (qr.data)
```
